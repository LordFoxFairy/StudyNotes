# 第1章 统计学习及监督学习概论

## 统计学习

### 什么是统计学习？
```
统计学习（统计机器学习）是关于计算基于数据构建概率统计模型并运用模型对数据进行预测与分析的一门学科。
```
### 基本分类

#### 监督学习

```
监督学习是指从标注数据中学习预测模型的机器学习问题。
```
实现步骤：
* 得到一个有限的训练数据集合
* 确定模型的假设空间，也就是所有的备选模型
* 确定模型选择的准则，即学习的策略
* 实现求解最优模型的算法
* 通过学习方法选择最优模型
* 利用学习的最优模型对新数据进行预测或分析

#### 无监督学习
```
无监督学习是指从无标注数据中的学习预测模型的机器学习问题。
```

#### 监督学习与无监督学习的区别
```
监督学习与无监督学习的区别是否有标签。
```

#### 强化学习
```
强化学习是指智能系统在与环境的连续互动中学习最优行为策略的机器学习问题。
```

#### 半监督学习
```
半监督学习是指利用标注数据和未标注数据学习预测模型的机器学习问题。
```

#### 主动学习
```
主动学习是指机器学习不断主动给出实例进行标注，然后利用标注数据学习预测模型的机器学习问题。
```
